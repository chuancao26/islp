# DecisionClassiiferTree

```python
carseats = load_data('Carseats')
carseats.head()
```

Vamos a hacer uso del clasificador del arbol simple, por lo que nos crearemos una variable de respuesta categorica. En este caso cuando los sales sean mayores a 8 seran high

```python
high = np.where(carseats["Sales"] > 8,
                "Yes",
                "No")
```

```python
model = MS(carseats.columns.drop("Sales"), intercept=False)
D = model.fit_transform(carseats)
feature_names = list(D.columns)
X = np.asarray(D)
X
```

Vamos a definir el clasificador


```python
clf = DecisionTreeClassifier(max_depth=3,
                             criterion="entropy",
                             random_state=0)
clf.fit(X, high)
```

```python
accuracy_score(high, clf.predict(X))
```
